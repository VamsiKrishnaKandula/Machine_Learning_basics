```python
"""Importing the modules"""
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Boston.csv')
data.head(30)
```

```python
"""check for the null values in the data"""
data.isnull().sum()
```

```python
"""The most important trick was to see the data correlation 
the relation between the dependent and independent columns
The correction will see with the 
-1 highly negatively correlated means they are oppposing more 
0 there is no correlation in between dependent and the independent terms
1 highly positively correlation there more correlation in between the dependent and the independent"""

data.corr()
```

```python
"""describing the data see which column has more std, mean, median, etc"""
data.describe()
```

```python
"""The most important was droppping the columns from the data
we think some columns are not used in the data 
we drop columns by see correlation tabel of the data
in the correlation tabel most of the coulns are negatively correlates
i will drop some columns by seeing data randomly
There is no use of the 
Unnamed: 0	
zn, chas"""

data = data.drop(columns = ['Unnamed: 0','zn','chas','black','tax'], axis = 0)
```

```python
data.head()
```

```python
"""NOW we have clean data i think 
NO NULL VALUES 
NO UNWANTED DATA COLUMNS"""

#FEATURES of the data
X = data.iloc[:, :-1].values
#ILOC WILL SEE THE LOCATION OF THE COLUMNS AND VALUES WILL BRING DATA FROM DATAFRAME TO NUMPY ARRAY
y = data.iloc[:, -1].values
```

```python
type(X) #this was changed from pandas to numpy
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.2)
```

```python
from sklearn.svm import SVR
svm = SVR(C = 2, kernel = 'rbf', degree = 2)
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```
